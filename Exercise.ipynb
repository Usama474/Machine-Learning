```python
from sklearn.datasets import load_digits

d= load_digits()
dir(d)
```

```python
import matplotlib.pyplot as plt
plt.gray()

for i in range(9):
    plt.matshow(d.images[i]) 
```

```python
import pandas as pd
df=X= pd.DataFrame(d.data, columns = d.feature_names)
df
```

```python
df['target']= d.target
df
```

```python
Y=df.target
Y
```

```python
plt.scatter(X.pixel_0_5, Y, color ='red')
plt.scatter(X.pixel_0_6, Y, color ='green')
plt.scatter(X.pixel_0_7, Y, color ='blue')
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest= train_test_split(X, Y, test_size=0.2)
```

```python
from sklearn.svm import SVC
model= SVC()
```

```python
model.fit(xtrain, ytrain, gamma=)
```

```python
model.score(xtest, ytest)
```
